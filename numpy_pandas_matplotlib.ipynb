```python
plt.plot(x,y, '-r')
plt.xlabel("x label")
plt.ylabel("Y label")
plt.title("Graph")
#plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b')
plt.subplot(1,2,2)
plt.plot(y,x,'y')
```

```python
#object oriented
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Graph")
```
